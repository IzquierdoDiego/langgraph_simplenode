```python
from typing import TypedDict

class PortfolioState(TypedDict):
    amount_usd: float
    total_usd: float
    total_inr: float
```

```python
def calc_total(state: PortfolioState) -> PortfolioState:
    state['total_usd'] = state['amount_usd'] * 80
    return state
```

```python
def calc_total_inr(state: PortfolioState) -> PortfolioState:
    state['total_inr'] = state['total_usd'] * 80
    return state
```

```python
from langgraph.graph import StateGraph, START, END

builder = StateGraph(PortfolioState)

builder.add_node("calc_total_node", calc_total)
builder.add_node("calc_total_inr_node", calc_total_inr)

builder.add_edge(START, "calc_total_node")
builder.add_edge("calc_total_node", "calc_total_inr_node")
builder.add_edge("calc_total_inr_node", END)

graph = builder.compile()
```

```python
from IPython.display import Image, display
from langchain_core.runnables.graph import MermaidDrawMethod

print(graph.get_graph().draw_mermaid())
```
